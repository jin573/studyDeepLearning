```python
# 다차원 배열을 지원하는 라이브러리
import numpy as np
print(np.__version__)
```

```python
# numpy로 배열 만들기 -> array 함수 사용
my_arr = np.array([[10, 20, 30], [40, 50, 60]])
print(my_arr)
```

```python
# 그래프 관련 라이브러리 맷플롯립
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25]) #x축 y축
plt.show()
```

```python
# 산점도 그리기
plt.scatter([1, 2, 3, 4, 5], [1, 4, 9, 16, 25]) #x축 y축
plt.show()
```

```python
# 랜덤 산점도 그리기
x = np.random.randn(1000) #평균이 0 이고 표준편자가 1인 정규분포난수
y = np.random.randn(1000)

plt.scatter(x, y) #x축 y축
plt.show()
```
